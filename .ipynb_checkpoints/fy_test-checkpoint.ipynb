### 特征5 时间段

```python
def get_time_interval(x):
    if len(x)!=19:return 0
    temp=int(x[11:13])
    if 1<=temp<=6:return 6
    elif 7<=temp<=12:return 12
    elif 13<=temp<=18:return 18
    elif 19<=temp<=20:return 20
    elif 21<=temp<=22:return 22
    else: return 24
```

```python
# 所属时间段：6为1-6；12为7-12；18为13-18；20为19-20；22为21-22；其他为24
%time metadata_mini['time_interval']=metadata_mini['time_contact'].apply(lambda x :get_time_interval(x))
```

```python
metadata_mini.head(3)
```

```python
sum_time=metadata_mini.groupby(['number','time_interval']).apply(lambda x:sum(x.num_contact)).to_frame()
# tt=pd.DataFrame(metadata_mini.groupby(['number','time_interval'],as_index=False).apply(lambda x:sum(x.num_contact)))
a_col=[]
b_col=[]
for i in sum_time.index:
    a_col.append(i[0])
    b_col.append(i[1])
sum_time['a_col']=a_col
sum_time['b_col']=b_col
aa=sum_time.pivot('a_col','b_col').fillna(0.0)
```

```python
sum_times=metadata_mini.groupby(['number','time_interval']).apply(lambda x:len(x)).to_frame()
a_col=[]
b_col=[]
for i in sum_times.index:
    a_col.append(i[0])
    b_col.append(i[1])
sum_times['a_col']=a_col
sum_times['b_col']=b_col
bb=sum_times.pivot('a_col','b_col').fillna(0.0)
```

## 制造特征234

```python
%%time
# time_stats={"total_time":np.sum,"avg_time":np.avg}
def get_stats(grouped,feature):
    stats=["sum","mean","count","std","max","min","median"]
    prefix=feature+"_"
    grouped=grouped[feature].agg(stats).add_prefix(prefix)
    return grouped
# call_contact_gap=app_call_record_group["contact_gap"].agg(get_stats).add_prefix("contact_gap_")
## 总计
app_call_record_group=metadata_mini.groupby(["number"])
print '排序已经完成'
record_time=get_stats(app_call_record_group,"num_contact")
print '总通话时常变量已经完成'
# record_contact_gap=get_stats(app_call_record_group,"contact_gap")
# print '通话间隔变量已经完成'
## 来电
app_call_record_group_BA=metadata_mini[metadata_mini.type_call==2].groupby(["number"])
print '来电排序已经完成'
record_time_BA=get_stats(app_call_record_group_BA,"num_contact").add_suffix('_come')
print '来电通话时常变量已经完成'
# record_contact_gap_BA=get_stats(app_call_record_group_BA,"contact_gap").add_suffix('_come')

## 去电
app_call_record_group_AB=metadata_mini[metadata_mini.type_call==1].groupby(["number"])
print '去电排序已经完成'
record_time_AB=get_stats(app_call_record_group_AB,"num_contact").add_suffix('_call_to')
print '去电通话时常变量已经完成'
# record_contact_gap_AB=get_stats(app_call_record_group_BA,"contact_gap").add_suffix('_call_to')


## 未接
# app_call_record_group_miss=app_call_record[app_call_record.type_call==3].groupby(["custorm_id","login_name","phone_contact"])["type"].count().add_suffix("_miss")


## 合并
```
