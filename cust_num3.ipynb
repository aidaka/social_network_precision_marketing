```python
records=sqlCtx.sql('''select number,-- 手机号
count(distinct custorm_id) as cust_num, -- 与其联系的用户个数
count(custorm_id) as freq_num,-- 用户与其联系总次数
sum(cast(time as float)) as total_time, --  用户与其联系总时间
avg(cast(time as float)) as avg_time, -- 用户与其联系平均时间
count(case when type="1" then number end) as  call_to_cust,-- 用户打给他的次数
sum(case when type="1" then time end) as  call_to_cust_total_time, -- 用户打给他总时长
sum(case when type="1" then time end)/(count(case when type="1" then number end)+1) as  call_to_cust_avg_time, -- 用户打给他平均时长
count(case when type="2" then number end) as  cust_to_him, -- 他打给用户的次数
sum(case when type="2" then time end) as  cust_to_him_total_time, -- 他打给用户总时长
sum(case when type="2" then time end)/(count(case when type="2" then number end)+1) as  cust_to_him_avg_time, --他打给用户平均时长
count(case when type="3" then number end) as  cust_miss_him, -- 用户未接的次数
sum(case when substring(call_time,12,2) between "01" and "06" then 1 end) as one_six, -- 1-6
sum(case when substring(call_time,12,2) between "01" and "06" then time end) as one_six_total_time,-- 1-6总时长
sum(case when substring(call_time,12,2) between "01" and "06" then time end)/(sum(case when substring(call_time,12,2) between "01" and "06" then 1 
end)+1) as one_six_avg_time,-- 1-6平均时长
sum(case when substring(call_time,12,2) between "07" and "12" then 1 end) as seven_tw, -- 7-12
sum(case when substring(call_time,12,2) between "07" and "12" then time end) as seven_tw_total_time,-- 7-12总时长
sum(case when substring(call_time,12,2) between "07" and "12" then time end)/(sum(case when substring(call_time,12,2) between "07" and "12" then 1 end)
 +1) as seven_tw_avg_time,-- 7-12平均时长
sum(case when substring(call_time,12,2) between "13" and "18" then 1 end) as thir_eig,  -- 13 -18
sum(case when substring(call_time,12,2) between "13" and "18" then time end) as thir_eig_total_time,-- 13-18总时长
sum(case when substring(call_time,12,2) between "13" and "18" then time end)/(sum(case when substring(call_time,12,2) between "13" and "18" then 1 end) 
+1) as thir_eig_avg_time,-- 13-18平均时长
sum(case when substring(call_time,12,2) between "19" and "20" then 1 end) as nin_twenty, -- 19-20
sum(case when substring(call_time,12,2) between "19" and "20" then time end) as nin_twenty_total_time,-- 19-20总时长
sum(case when substring(call_time,12,2) between "19" and "20" then time end)/(sum(case when substring(call_time,12,2) between "19" and "20" then 1 end)
+1) as nin_twenty_avg_time,-- 19-20平均时长
sum(case when substring(call_time,12,2) between "21" and "22" then 1 end) as twenty_one_two, -- 21-22
sum(case when substring(call_time,12,2) between "21" and "22" then time end) as twenty_one_two_total_time,-- 21-22总时长
sum(case when substring(call_time,12,2) between "21" and "22" then time end)/(sum(case when substring(call_time,12,2) between "21" and "22" then 1 end)
+1) as twenty_one_two_avg_time,-- 21-22平均时长
sum(case when substring(call_time,12,2) in ("23","24","00")  then 1 end)  as sleep, -- 23，24，00 
sum(case when substring(call_time,12,2) in ("23","24","00")  then time end) as sleep_total_time,-- 23，24，00 总时长
sum(case when substring(call_time,12,2) in ("23","24","00")  then time end)/(sum(case when substring(call_time,12,2) in ("23","24","00")  then 1 end) 
+1) as sleep_avg_time -- 23，24，00 平均时长
 from   crawler.app_call_record_hive
 where  cast(number as float)>13000000000
and cast(number as float)<19000000000
group by number,name  having cust_num=3
''')
records=records.drop_duplicates?
```
